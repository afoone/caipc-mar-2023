```python
import os
import skimage
import matplotlib.pyplot as plt
```

```python
def load_data(data_dir): 
    dirs = os.listdir(data_dir)
    images = []
    labels = []
    for d in dirs:
        relative_path = os.path.join(data_dir, d)
        file_names = [
            os.path.join(relative_path, f)
            for f in os.listdir(relative_path)
            if f.endswith('.ppm')
        ]
        for f in file_names:
            images.append(skimage.io.imread(f))
            labels.append(int(d))
    return images, labels
            
            
    
    
```

```python
X, y = load_data('./Training')
```

```python
from skimage.transform import resize
from skimage.color import rgb2gray



```

```python
# reescalado
Xr = [
    resize(img, (30,30))
    for img in X
]
```

```python
X = [
    rgb2gray(img)
    for img in Xr
]
```

```python
import tensorflow as tf
```

```python
# modelo secuencial
model = tf.keras.models.Sequential()
```

```python
# vamos a pasar de un shape de (30,30) a uno de (900)
model.add(tf.keras.layers.Flatten(input_shape=(30,30)))
```

```python
# vamos a añadir una fila de neuronas con X neuronas y una función de activación 
model.add(tf.keras.layers.Dense(9, activation='sigmoid'))
```

```python
# regularización de datos
model.add(tf.keras.layers.Dropout(0.2))
```

```python
# salida
model.add(tf.keras.layers.Dense(62, activation='softmax'))
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics="accuracy"
)
```

```python
t_X = tf.constant(X)
```

```python
t_y = tf.constant(y)
```

```python
len(t_y)
```

```python
model.fit(
    t_X,
    t_y,
    epochs=50
)
```
